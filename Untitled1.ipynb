```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from google.colab import files
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.optimizers import Adam # Import Adam optimizer
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.optimizers import Adam


# Upload the train and test files
print("Please upload 'train.csv' and 'test.csv' when prompted.")
try:
    uploaded = files.upload()  # train.csv
    uploaded2 = files.upload()  # test.csv
except Exception as e:
    print(f"Error during file upload: {e}. Please ensure 'train.csv' and 'test.csv' are uploaded.")
    raise

# Load data
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")

# Display first 5 rows
print("First 5 rows of training data:")
display(train_df.head())

# Check missing values
print("\nMissing values in training data (before dropping high missing % columns):")
display(train_df.isnull().sum()[train_df.isnull().sum() > 0])

# --- NEW: Drop columns with more than 75% missing values ---
def drop_high_missing_cols(df, threshold=0.75):
    """
    Drops columns from a DataFrame that have a missing value percentage
    greater than the specified threshold.
    """
    missing_percentages = df.isnull().sum() / len(df)
    cols_to_drop = missing_percentages[missing_percentages > threshold].index.tolist()
    if cols_to_drop:
        print(f"\nDropping columns with >{threshold*100}% missing values: {cols_to_drop}")
        df = df.drop(columns=cols_to_drop)
    else:
        print(f"\nNo columns found with >{threshold*100}% missing values to drop.")
    return df, cols_to_drop

# Apply to a copy of both train and test datasets to determine all columns to drop
train_df_temp = train_df.copy()
test_df_temp = test_df.copy()

train_df_temp, dropped_from_train = drop_high_missing_cols(train_df_temp, threshold=0.75)
test_df_temp, dropped_from_test = drop_high_missing_cols(test_df_temp, threshold=0.75)

all_dropped_high_missing = list(set(dropped_from_train + dropped_from_test))
if all_dropped_high_missing:
    train_df.drop(columns=all_dropped_high_missing, inplace=True, errors='ignore')
    test_df.drop(columns=all_dropped_high_missing, inplace=True, errors='ignore')
    print(f"\nEnsured consistent dropping of high missing value columns across both datasets: {all_dropped_high_missing}")


# Fill missing values with median for numeric and mode for categorical
def fill_missing(df):
    """
    Fills missing values in a DataFrame:
    - Numeric columns: filled with median.
    - Categorical columns: filled with mode.
    """
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype == 'object':
                df[col] = df[col].fillna(df[col].mode()[0])
            else:
                df[col] = df[col].fillna(df[col].median())
    return df

train_df = fill_missing(train_df)
test_df = fill_missing(test_df)

print("\nâœ… Missing values filled using median/mode!")

# --- NEW: Target Variable Transformation (Log Transform SalePrice) ---
# Apply log1p transformation to SalePrice in training data
# This helps normalize the distribution and can improve model performance
train_df['SalePrice'] = np.log1p(train_df['SalePrice'])
print("\nâœ… SalePrice log-transformed!")

# Remove outliers from training data only (not test) - MODIFIED to cap outliers instead of removing rows
def cap_outliers(df, columns):
    """
    Caps outliers in specified numeric columns using the IQR method.
    Values below lower bound are set to lower bound, values above upper bound are set to upper bound.
    """
    df_copy = df.copy() # Work on a copy
    for col in columns:
        if col in df_copy.columns and pd.api.types.is_numeric_dtype(df_copy[col]):
            Q1 = df_copy[col].quantile(0.25)
            Q3 = df_copy[col].quantile(0.75)
            IQR = Q3 - Q1
            lower = Q1 - 1.5 * IQR
            upper = Q3 + 1.5 * IQR
            df_copy[col] = np.where(df_copy[col] < lower, lower, df_copy[col])
            df_copy[col] = np.where(df_copy[col] > upper, upper, df_copy[col])
    print(f"Capped outliers in numeric columns.")
    return df_copy

num_cols_for_capping = train_df.select_dtypes(include=[np.number]).columns.tolist()
if 'Id' in num_cols_for_capping: num_cols_for_capping.remove('Id')
if 'SalePrice' in num_cols_for_capping: num_cols_for_capping.remove('SalePrice') # Do not cap SalePrice after log transform, as it's the target

train_df = cap_outliers(train_df, num_cols_for_capping)
print("Shape after capping outliers (no rows removed):", train_df.shape)


# --- CRITICAL CHANGE: Store test_ids BEFORE dropping 'Id' from test_df ---
test_ids = test_df['Id'] # Store 'Id' column from original test_df

# Drop irrelevant and mostly missing columns (adjusted based on new high missing % drop)
# 'Id' is handled separately for test_df
drop_cols_from_features = ['Utilities', 'LowQualFinSF', '3SsnPorch',
                           'ScreenPorch', 'BsmtHalfBath', 'KitchenAbvGr', 'GarageYrBlt']

# Filter out columns that were already dropped by the 75% rule
drop_cols_from_features = [col for col in drop_cols_from_features if col not in all_dropped_high_missing]

# Drop from train_df
train_df.drop(columns=[col for col in drop_cols_from_features if col in train_df.columns], inplace=True, errors='ignore')

# Drop from test_df (excluding 'Id' here, as it's handled by test_ids)
test_df.drop(columns=[col for col in drop_cols_from_features if col in test_df.columns], inplace=True, errors='ignore')
print(f"\nDropped specified irrelevant columns from features: {drop_cols_from_features}")


# --- NEW: Feature Engineering ---
def create_features(df):
    df_copy = df.copy()

    # Total Square Footage - Safely get columns with default 0
    df_copy['TotalSF'] = df_copy.get('GrLivArea', 0) + df_copy.get('TotalBsmtSF', 0)

    # Total Bathrooms - Safely get columns with default 0
    df_copy['TotalBath'] = (df_copy.get('FullBath', 0) + (df_copy.get('HalfBath', 0) * 0.5) +
                             df_copy.get('BsmtFullBath', 0) + (df_copy.get('BsmtHalfBath', 0) * 0.5))

    # Age of House at Sale - Safely get columns with default 0
    df_copy['AgeAtSale'] = df_copy.get('YrSold', 0) - df_copy.get('YearBuilt', 0)
    df_copy['AgeAtSale'] = df_copy['AgeAtSale'].apply(lambda x: max(0, x)) # Ensure no negative ages

    # Years Since Remodel - Safely get columns with default 0
    df_copy['YearsSinceRemodel'] = df_copy.get('YrSold', 0) - df_copy.get('YearRemodAdd', 0)
    df_copy['YearsSinceRemodel'] = df_copy['YearsSinceRemodel'].apply(lambda x: max(0, x)) # Ensure no negative values

    # Overall House Quality Interaction - Safely get columns with default 1 to avoid multiplying by 0
    df_copy['OverallHouseQuality'] = df_copy.get('OverallQual', 1) * df_copy.get('OverallCond', 1)

    # Binary features for presence of amenities - Safely get columns with default 0
    df_copy['HasPool'] = (df_copy.get('PoolArea', 0) > 0).astype(int)
    df_copy['HasGarage'] = (df_copy.get('GarageArea', 0) > 0).astype(int)
    df_copy['HasBsmt'] = (df_copy.get('TotalBsmtSF', 0) > 0).astype(int)
    df_copy['HasFireplace'] = (df_copy.get('Fireplaces', 0) > 0).astype(int)

    return df_copy

train_df = create_features(train_df)
test_df = create_features(test_df)
print("\nâœ… New features engineered!")


# Separate target
y = train_df['SalePrice']
X = train_df.drop('SalePrice', axis=1)

# Drop 'Id' from X and test_df_for_processing before concatenation
if 'Id' in X.columns:
    X = X.drop(columns=['Id'])
if 'Id' in test_df.columns: # test_df still contains 'Id' at this point
    test_df_for_processing = test_df.drop(columns=['Id'], errors='ignore')
else:
    test_df_for_processing = test_df.copy() # If 'Id' was already dropped, just copy

print(f"\nShape of X (features): {X.shape}")
print(f"Shape of y (target): {y.shape}")
if X.shape[0] != y.shape[0]:
    raise ValueError("Mismatch in number of rows between features (X) and target (y) after preprocessing.")


# --- REVISED SECTION: Combine for consistent one-hot encoding and then split ---
n_train_rows = X.shape[0]
n_test_rows = test_df_for_processing.shape[0]

combined_features = pd.concat([X, test_df_for_processing], ignore_index=True)

# Perform one-hot encoding on the entire combined features DataFrame
combined_features_encoded = pd.get_dummies(combined_features)

# Split the one-hot encoded combined DataFrame back into processed train and test sets
X_processed = combined_features_encoded.iloc[:n_train_rows]
test_processed = combined_features_encoded.iloc[n_train_rows:]

# Ensure both X_processed and test_processed have the same columns and order
all_columns = X_processed.columns.union(test_processed.columns)
X_processed = X_processed.reindex(columns=all_columns, fill_value=0)
test_processed = test_processed.reindex(columns=all_columns, fill_value=0)

X_processed = X_processed.sort_index(axis=1)
test_processed = test_processed.sort_index(axis=1)

# Print one-hot encoded data info
print("\nðŸ”¢ One-Hot Encoded Data (after pd.get_dummies and splitting):")
print("Shape of X_processed:", X_processed.shape)
display(X_processed.head())
print("Shape of test_processed:", test_processed.shape)
display(test_processed.head())

# Verify column count consistency
if X_processed.shape[1] != test_processed.shape[1]:
    raise ValueError(f"Column mismatch after one-hot encoding. X_processed has {X_processed.shape[1]} columns, test_processed has {test_processed.shape[1]} columns.")
else:
    print(f"âœ… Column count consistent: {X_processed.shape[1]} columns in both X_processed and test_processed.")


# Scale the features and convert to float32
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_processed).astype(np.float32)
test_scaled = scaler.transform(test_processed).astype(np.float32)

# Verify shapes and types after scaling
print(f"Shape of X_scaled: {X_scaled.shape}, Dtype: {X_scaled.dtype}")
print(f"Shape of test_scaled: {test_scaled.shape}, Dtype: {test_scaled.dtype}")


# Plot using scatter plots (using the example data as before)
example_data = {
    'LotArea': [8450, 9600, 11250, 9550, 14260],
    'OverallQual': [7, 6, 7, 7, 8],
    'YearBuilt': [2003, 1976, 2001, 1915, 2000]
}
df_example = pd.DataFrame(example_data, index=[f'Row {i+1}' for i in range(5)])

scaler_example = MinMaxScaler()
scaled_array_example = scaler_example.fit_transform(df_example)
scaled_df_example = pd.DataFrame(scaled_array_example, columns=df_example.columns, index=df_example.index)

plt.figure(figsize=(14, 5))
for i, column in enumerate(df_example.columns):
    plt.subplot(1, 3, i+1)
    plt.scatter(df_example.index, df_example[column], label='Before Scaling', color='blue', s=80)
    plt.scatter(df_example.index, scaled_df_example[column], label='After Scaling', color='orange', marker='x', s=80)
    plt.title(column)
    plt.xticks(rotation=45)
    plt.ylabel('Value')
    plt.legend()
plt.suptitle('ðŸ“‰ Scatter Plot: Feature Values Before vs After Scaling (Example Data)', fontsize=14)
plt.tight_layout()
plt.show()

from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.optimizers import Adam

# Split data
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Build improved model
model = models.Sequential([
    layers.Input(shape=(X_train.shape[1],)),

    layers.Dense(2048),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.4),

    layers.Dense(1024),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.3),

    layers.Dense(512),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.2),

    layers.Dense(256),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.1),

    layers.Dense(64),
    layers.BatchNormalization(),
    layers.Activation('relu'),

    layers.Dense(1)  # Output
])

# Compile model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

# Callbacks
early_stop = callbacks.EarlyStopping(patience=50, restore_best_weights=True, monitor='val_loss')
reduce_lr = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=20, min_lr=1e-5, verbose=1)

# Train model
history = model.fit(X_train, y_train,
                    validation_data=(X_val, y_val),
                    epochs=500,
                    batch_size=32,
                    callbacks=[early_stop, reduce_lr],
                    verbose=1)

# Predict and evaluate
y_pred_transformed = model.predict(X_val).flatten()
y_pred = np.expm1(y_pred_transformed) # Inverse transform predictions
y_val_original = np.expm1(y_val) # Inverse transform actual values for evaluation

mse = mean_squared_error(y_val_original, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_val_original, y_pred)

# Print first 5 actual vs predicted SalePrice
comparison_df = pd.DataFrame({
    'Actual SalePrice': y_val_original.values[:5],
    'Predicted SalePrice': y_pred[:5]
})
print("\n First 5 Actual vs Predicted SalePrice (Inverse Transformed):")
display(comparison_df)

print("\nðŸ“Š Evaluation Metrics:")
print("ðŸ”¹ MSE:", mse)
print("ðŸ”¹ RMSE:", rmse)
print("ðŸ”¹ RÂ² Score (Accuracy):", r2)

# Final prediction on test set
test_predictions_transformed = model.predict(test_scaled).flatten()
test_predictions = np.expm1(test_predictions_transformed) # Inverse transform test predictions

# Plot training history
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train MSE')
plt.plot(history.history['val_loss'], label='Val MSE')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.title('Loss Over Epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Val MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('MAE Over Epochs')
plt.legend()

plt.tight_layout()
plt.show()

# Save test predictions
submission = pd.DataFrame({'Id': test_ids, 'SalePrice': test_predictions})
submission.to_csv("submission.csv", index=False)
print("\nâœ… Predictions saved to submission.csv")
```
